#Theoritical Questions

**1. What is Machine Learning? Explain the main types of Machine Learning with examples.**

  = Machine Learning is the study of algorithms that allow computers to learn from experience (data) and make predictions or decisions automatically.


* Supervised learning

  : In supervised learning, the algorithm is trained using labeled data, where the correct output is already known.  Ex. : House price prediction
* Unsupervised learning

  : In unsupervised learning, the algorithm works with unlabeled data. The model tries to discover hidden patterns or structures in the data. Ex. : Grouping similar news articles
* reinforcement learning

  : In reinforcement learning, an agent learns by interacting with an environment and receiving rewards or penalties. Ex. : Self-driving cars

**2. What is Supervised Learning? Explain the difference between Regression and Classification.**

  =  Supervised Learning is a type of machine learning in which the model is trained using labeled data. Each training example consists of an input and its correct output (label). The algorithm learns the relationship between the input and output so it can predict the correct output for new, unseen data.



* Regression : Used to predict continuous numerical values, such as prices, temperatures, or salaries.
* Classification: Used to predict categorical Values, such as Spam/Not Spam, Pass/Fail, or Positive/Negative.



**Q3. What is Linear Regression? Write the Linear Regression equation and explain intercept and coefficient.**

   = Linear Regression is a supervised machine learning algorithm used to predict a continuous numerical value by finding the best-fit straight line that describes the relationship between an independent variable (X) and a dependent variable (Y).

  Equation : y = b₀ + b₁x



* Intercept (b₀) : The intercept is the value of the dependent variable (Y) when the independent variable (X) is 0.
* Coefficient (b₁) : The coefficient indicates how much the predicted value (Y) changes for every one-unit increase in X.



**Q4. Explain the following evaluation metrics used in Linear Regression:**



  

* SSE : sum of square error is the sum of the squared differences between actual and predicted values.

* MSE : mean square error is the average of the squared errors.

* RMSE : Root mean square error is the square root of the MSE.
* R² SCORE : R² measures how well the model explains the variation in the target variable.


* Adjusted R² : Adjusted R² is a modified version of R² that considers the number of independent variables in the model.

**Q8. Select independent variables (features) and dependent variable (target).
Target variable: medv (house price).**

```python
# Independent variables (Features)
X = df.drop('medv', axis=1)

# Dependent variable (Target)
y = df['medv']

# Display features and target
print("Independent Variables (X):")
print(X.head())

print("\nDependent Variable (y):")
print(y.head())
```

**Q9. Split the dataset into training and testing sets (80% training, 20% testing).**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42
)

# Display the shape of each dataset
print("X_train shape:", X_train.shape)
print("X_test shape :", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape :", y_test.shape)
```

**Q10. Train a Linear Regression model and perform the following:**

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
```

```python
model = LinearRegression()

model.fit(X_train, y_train)
# Predict house prices
y_pred = model.predict(X_test)

print(y_pred[:10])
```

```python
#MAE
mae = mean_absolute_error(y_test, y_pred)
print("MAE:", mae)
```

```python
#MSE
mse = mean_squared_error(y_test, y_pred)
print("MSE:", mse)
```

```python
#RMSE
rmse = np.sqrt(mse)
print("RMSE:", rmse)
```

```python
#R² Score
r2 = r2_score(y_test, y_pred)
print(r2)
```

```python
#Adjusted R²
n = X_test.shape[0]      # Number of observations
p = X_test.shape[1]      # Number of features

adjusted_r2 = 1 - ((1-r2)*(n-1)/(n-p-1))

print(adjusted_r2)
```

```python
#Linear regression equation
print("Intercept (b0):", model.intercept_)

print("\nCoefficients:")
for feature, coef in zip(X.columns, model.coef_):
    print(feature, ":", coef)
```

```python
#interpret the model coefficient
coef_df = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': model.coef_
})

print(coef_df)
```
